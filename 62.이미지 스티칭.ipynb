```python
img_names = ['img1.jpg', 'img2.jpg', 'img3.jpg']


imgs = []

for name in img_names:#img_names  name 

    img = cv2.imread(name) #name  img 



    if img is None:

        print('Image load failed!')

        sys.exit()


    imgs.append(img) 
```

```python
plt.subplot(131),plt.axis('off'),plt.imshow(imgs[0])
plt.subplot(132),plt.axis('off'),plt.imshow(imgs[1])
plt.subplot(133),plt.axis('off'),plt.imshow(imgs[2])
```

#   

```python
stitcher = cv2.Stitcher_create()
```

#      . 

```python
status, dst = stitcher.stitch(imgs)
```

```python
if status != cv2.Stitcher_OK:

    print('Stitch failed!')

    sys.exit()
```

```python
cv2.imwrite('output.jpg', dst)
```
