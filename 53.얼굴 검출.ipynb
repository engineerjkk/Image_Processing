```python
src = cv2.imread('lenna.bmp')
```

```python
if src is None:
    print('Image load failed!')
    sys.exit()
```

   .

```python
classifier = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')
```

```python
if classifier.empty():
    print('XML load failed!')
    sys.exit()
# 
tm=cv2.TickMeter()
tm.start()

faces = classifier.detectMultiScale(src,scaleFactor=2)
tm.stop()
print(tm.getTimeMilli())
print(len(faces))
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(src, (x, y, w, h), (255, 0, 255), 2)
```

```python
src2=cv2.cvtColor(src,cv2.COLOR_RGB2BGR)
plt.axis('off'),plt.imshow(src2)
```
